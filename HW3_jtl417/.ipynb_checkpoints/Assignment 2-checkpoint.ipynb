**IDEA:**
In January 2015, people were more likely to use citi bike on weekends. 

**Null Hypothesis**

In words: In January 2015, The number of trips made per day on the weekends (e.g. (# trips on Saturday/Sunday)/2) is less than or equal to the number of trips made per day on weekdays (e.g. (# trips on Monday/Tuesday/Wednesday/Thursday/Friday)/5).

As a formula: (# trips on (Saturday/Sunday)/2) <= (# trips on Monday/Tuesday/Wednesday/Thursday/Friday)/5)

**Alternative Hypothesis**

In words: In January 2015, The number of trips made per day on the weekends is greater than the number of trips made per day on weekdays.

As a formula: (# trips on Saturday/Sunday)/2) > (# trips on Monday/Tuesday/Wednesday/Thursday/Friday)/5)

**CONFIDENCE LEVEL: ALPHA = .05**

```python
import os
import pylab as pl
import seaborn
import pandas as pd
import numpy as np
from datetime import datetime as dt
%pylab inline
```

```python
def getCitiBikeCSV(datestring):
    print ("Downloading", datestring)
    ### First I will heck that it is not already there
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        if os.path.isfile(datestring + "-citibike-tripdata.csv"):
            # if in the current dir just move it
            if os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA")):
                print ("Error moving file!, Please check!")
        #otherwise start looking for the zip file
        else:
            if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip"):
                if not os.path.isfile(datestring + "-citibike-tripdata.zip"):
                    os.system("curl -O https://s3.amazonaws.com/tripdata/" + datestring + "-citibike-tripdata.zip")
                ###  To move it I use the os.system() functions to run bash commands with arguments
                os.system("mv " + datestring + "-citibike-tripdata.zip " + os.getenv("PUIDATA"))
            ### unzip the csv 
            os.system("unzip " + os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.zip")
            ## NOTE: old csv citibike data had a different name structure. 
            if '2014' in datestring:
                os.system("mv " + datestring[:4] + '-' +  datestring[4:] + 
                          "\ -\ Citi\ Bike\ trip\ data.csv " + datestring + "-citibike-tripdata.csv")
            os.system("mv " + datestring + "-citibike-tripdata.csv " + os.getenv("PUIDATA"))
    ### One final check:
    if not os.path.isfile(os.getenv("PUIDATA") + "/" + datestring + "-citibike-tripdata.csv"):
        print ("WARNING!!! something is wrong: the file is not there!")

    else:
        print ("file in place, you can continue")
```

```python
datestring = '201501'
getCitiBikeCSV(datestring)
```

```python
trips = pd.read_csv(os.getenv("PUIDATA") + "/" + datestring + '-citibike-tripdata.csv')
# df is the dataframe where the content of the csv file is stored
trips['date'] = pd.to_datetime(trips['starttime'])
# note that with dataframes I can refer to variables as dictionary keys, 
# i.e. df['starttime'] or as attributes: df.starttime. 
trips.head()
```

```python
trips['DayofWeek'] = trips['date'].dt.weekday

weekends = trips.groupby('DayofWeek').get_group(5)
weekends = weekends.append(trips.groupby('DayofWeek').get_group(6))

weekdays = trips.groupby('DayofWeek').get_group(0)
weekdays = weekdays.append(trips.groupby('DayofWeek').get_group(1))
weekdays = weekdays.append(trips.groupby('DayofWeek').get_group(2))
weekdays = weekdays.append(trips.groupby('DayofWeek').get_group(3))
weekdays = weekdays.append(trips.groupby('DayofWeek').get_group(4))
```

```python
weekdays.head()
```

```python
plt.hist(weekdays.DayofWeek)
plt.title('Histogram of Trips Taken by Weekend Day')
plt.xlabel("Day of Week")
plt.ylabel("N")
plt.xticks(range(0,5), ('Monday', "Tuesday", "Wednesday", "Thursday", "Friday"))
```
