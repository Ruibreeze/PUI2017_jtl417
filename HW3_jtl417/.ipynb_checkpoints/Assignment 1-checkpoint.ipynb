```python
import pylab as pl
import seaborn
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit, minimize
import scipy.stats as stats
%pylab inline
```

```python
#Initializing import objects
np.random.seed(14)
mean = 10
std = 5
norm = []
uni = []
pois = []
bi = []
chi = []
```

```python
#Creating 100 sets of samples that are normally distributed
n = 10
for i in range(0,100):
    sample = np.random.normal(mean, std, n)
    norm.append(sample)
    n += 15
```

```python
#Creating 100 sets of samples that are uniformally distributed
n = 10
for i in range(0,100):
    sample = np.random.uniform(low = mean - 10, high = mean + 10, size = (n))
    uni.append(sample)
    n += 15
```

```python
#Creating 100 sets of samples that are of the poisson distribution
n = 10
for i in range(0,100):
    sample = np.random.poisson(lam = mean, size = (n))
    pois.append(sample)
    n += 15
```

```python
#Creating 100 sets of samples that are binomially distributed
n = 10
for i in range(0,100):
    sample = np.random.binomial(n = (mean*10), p = .1, size = (n))
    bi.append(sample)
    n += 15
```

```python
#Creating 100 sets of samples that are of the chi square distribution
n = 10
for i in range(0,100):
    sample = np.random.chisquare(mean, size = (n))
    chi.append(sample)
    n += 15
```

```python
#Merging into a dictionary organizing samples by distribution (Distribution Type is key)
distributions = {'normal' : norm, 'uniform' : uni, 'possion' : pois, 'binomial' : bi, 'chi-squared' : chi}
```

```python
#Calculating + Plotting Sample Means
f, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex='col', sharey='row', figsize = (20,10))

samplemeans_tot = []
samplesize_tot = []

sample_means = []
sample_size = []
for i in range(0,len(norm)):
    sample = norm[i]
    size = len(sample)
    mean = np.mean(sample)
    sample_means.append(mean)
    sample_size.append(size)
ax1.scatter(sample_means,sample_size)
ax1.set_xlabel('Sample Mean')
ax1.set_ylabel('Sample Size')
ax1.set_title("Normal")
samplemeans_tot.append(sample_means)
samplesize_tot.append(sample_size)

sample_means = []
sample_size = []
for i in range(0,len(uni)):
    sample = uni[i]
    size = len(sample)
    mean = np.mean(sample)
    sample_means.append(mean)
    sample_size.append(size)
ax2.scatter(sample_means,sample_size)
ax2.set_xlabel('Sample Mean')
ax2.set_ylabel('Sample Size')
ax2.set_title("Uniform")
samplemeans_tot.append(sample_means)
samplesize_tot.append(sample_size)


sample_means = []
sample_size = []
for i in range(0,len(pois)):
    sample = pois[i]
    size = len(sample)
    mean = np.mean(sample)
    sample_means.append(mean)
    sample_size.append(size)
ax3.scatter(sample_means,sample_size)
ax3.set_xlabel('Sample Mean')
ax3.set_ylabel('Sample Size')
ax3.set_title("Poisson")
samplemeans_tot.append(sample_means)
samplesize_tot.append(sample_size)

sample_means = []
sample_size = []
for i in range(0,len(bi)):
    sample = bi[i]
    size = len(sample)
    mean = np.mean(sample)
    sample_means.append(mean)
    sample_size.append(size)
ax4.scatter(sample_means,sample_size)
ax4.set_xlabel('Sample Mean')
ax4.set_ylabel('Sample Size')
ax4.set_title("Binomial")
samplemeans_tot.append(sample_means)
samplesize_tot.append(sample_size)

sample_means = []
sample_size = []
for i in range(0,len(chi)):
    sample = chi[i]
    size = len(sample)
    mean = np.mean(sample)
    sample_means.append(mean)
    sample_size.append(size)
ax5.scatter(sample_means,sample_size)
ax5.set_xlabel('Sample Mean')
ax5.set_ylabel('Sample Size')
ax5.set_title("Chi-Squared")
samplemeans_tot.append(sample_means)
samplesize_tot.append(sample_size)

ax7.hist(samplemeans_tot, bins = 30)
ax7.set_xlabel("Mean")
ax7.set_ylabel("N")
ax7.set_title("Histogram of Sample Means")
```
